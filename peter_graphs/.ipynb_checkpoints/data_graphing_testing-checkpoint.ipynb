# Veronika Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
base_path = r'C:\Users\Peter\Downloads\2DG nof2.xlsx'
df = pd.read_excel(base_path)
```

```python
df_cntrl = df[df['trt'] == 'Control']
df_2dg = df[df['trt'] == '2DG']

df_1 = df[df['rep'] == 1]
df_1_cntrl = df_1[df_1['trt'] == 'Control']
df_1_2dg = df_1[df_1['trt'] == '2DG']

df_2 = df[df['rep'] == 2]
df_2_cntrl = df_2[df_2['trt'] == 'Control']
df_2_2dg = df_2[df_2['trt'] == '2DG']
```

```python
sns.set(rc={'figure.figsize':(15,15)})
sns.set_style(style='white')
```

```python
palette = {'Control': 'w', '2DG':'w'}
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
fig.suptitle("Redox Ratio", fontsize=50, weight='bold')
x_axis = 'trt'
y_axis = 'rr'
y_label = "Redox Ratio"
x_labels = ["Control", "2DG"]

PROPS_BLACK = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'},
    'capprops':{'color':'black'}
}
red_val = 135
PROPS_RED = {
    'boxprops':{'facecolor':'none', 'edgecolor':(red_val/255,  0, 0)},
    'medianprops':{'color':(red_val/255,  0, 0)},
    'whiskerprops':{'color':(red_val/255,  0, 0)},
    'capprops':{'color':(red_val/255, 0, 0)}
}
sns.boxplot(x=x_axis, y=y_axis, data=df_cntrl, palette=palette, ax=ax1, linewidth=2, **PROPS_BLACK)
sns.boxplot(x=x_axis, y=y_axis, data=df_2dg, palette=palette, ax=ax2, linewidth=5, **PROPS_RED)
sns.swarmplot(x=x_axis, y=y_axis, data=df_1_cntrl, color='.25', size=10, ax=ax1, dodge=True)
sns.swarmplot(x=x_axis, y=y_axis, data=df_1_2dg, color='r', size=20,ax=ax2, dodge=True)
sns.swarmplot(x=x_axis, y=y_axis, data=df_2_cntrl, color='0.75', size=10,ax=ax1, dodge=True)
sns.swarmplot(x=x_axis, y=y_axis, data=df_2_2dg, color=(255/255, 204/255, 203/255), size=10, ax=ax2, dodge=True)
ax1.set_yticklabels([str(round(i, 2)) for i in ax1.get_yticks()], fontsize=20, weight='bold')
ax1.set_xticklabels([str(i) for i in ax1.get_xticks()], fontsize=0)
ax2.set_xticklabels([str(i) for i in ax2.get_xticks()], fontsize=0)
ax1.set_xlabel(x_labels[0], fontsize=30, weight='bold')
#ax1.set_ylim(0, 1)
ax2.set_xlabel(x_labels[1], fontsize=30, weight='bold')
ax1.set_ylabel(y_label, fontsize=30, weight='bold')
ax2.set_ylabel(y_label, fontsize=0, weight='normal')
#ax1.set_yticks(list(np.linspace(0, 1, num=5)))
#ax1.set_yticklabels(list(np.linspace(0, 1, num=5)))
sns.despine(left=False, bottom=False, right=True, ax=ax1)
sns.despine(left=True, bottom=False, right=True, ax=ax2)
plt.rcParams['svg.fonttype'] = 'none'
fig.savefig('redox_ratio_demo_fig.svg')
```
