# Veronika Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
base_path = r'C:\Users\Peter\Downloads\2DG nof2.xlsx'
df = pd.read_excel(base_path)
```

```python
sns.set(rc={'figure.figsize':(15,15)})
```

```python
ax = sns.boxplot(x = "trt", y = "rr", data = df)
ax = sns.swarmplot(x="trt", y="rr", data=df, color=".25")
```

```python
ax = sns.boxplot(x = "trt", y = "ntm", data = df)
ax = sns.swarmplot(x="trt", y="ntm", data=df, color=".25")
```

```python
ax = sns.boxenplot(x="trt", y="rr", data=df,
                   showfliers=False, k_depth='tukey')
ax = sns.stripplot(x="trt", y="rr", data=df,
                   size=4, color=".26")
```
