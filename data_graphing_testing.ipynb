```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
base_path = r'C:\Users\prehani\Downloads\2DG nof2.xlsx'
df = pd.read_excel(base_path)
```

```python
ax = sns.boxplot(x = "fish", y = "rr", data = df)
ax = sns.swarmplot(x="fish", y="rr", data=df, color=".25")
```
